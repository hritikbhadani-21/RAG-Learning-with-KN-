```python
from langchain_community.document_loaders import WebBaseLoader
import bs4

loader=WebBaseLoader(web_path="https://medium.com/data-science-in-your-pocket/attention-is-all-you-need-understanding-with-example-c8d074c37767",
                     bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                         class_=("post-title","post-content","post-header")
                     )))
text_document=loader.load()
```

```python
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader("attension.pdf")
pdf_document=loader.load()
pdf_document
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
documents=text_splitter.split_documents(pdf_document)
pdf_document[:5]
```

```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import OllamaEmbeddings

embeddings = OllamaEmbeddings(model="nomic-embed-text")

db = Chroma.from_documents(pdf_document[:20], embeddings)
```

```python
query="what is attention is all you need"
result=db.similarity_search(query)
result
```
